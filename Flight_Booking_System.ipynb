```python
import csv
import os
import random
from datetime import datetime
import time

FLIGHT_FIELDS = ['flight_id', 'origin', 'destination', 'date', 'time', 'price', 'airline', 'seats']
BOOKING_FIELDS = ['booking_id', 'username', 'flight_id', 'seat', 'timestamp', 'amount', 'payment_ref']
USER_FIELDS = ['username', 'password']
FLIGHTS_CSV = 'flights.csv'
BOOKINGS_CSV = 'bookings.csv'
USERS_CSV = 'users.csv'

def load_csv(filename, fieldnames):
    if not os.path.exists(filename):
        with open(filename, 'w', newline='') as f:
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
        return []
    with open(filename, 'r', newline='') as f:
        return [row for row in csv.DictReader(f)]

def save_csv(filename, fieldnames, data):
    with open(filename, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for row in data:
            writer.writerow(row)

def register_user():
    users = load_csv(USERS_CSV, USER_FIELDS)
    username = input("Choose a username: ")
    if any(u['username'] == username for u in users):
        print("Username already exists.")
        return None
    password = input("Choose a password: ")
    users.append({'username': username, 'password': password})
    save_csv(USERS_CSV, USER_FIELDS, users)
    print("Registration successful!")
    return username

def login_user():
    users = load_csv(USERS_CSV, USER_FIELDS)
    username = input("Username: ")
    password = input("Password: ")
    for user in users:
        if user['username'] == username and user['password'] == password:
            print("Login successful!")
            return username
    print("Invalid credentials.")
    return None

def load_flights():
    return load_csv(FLIGHTS_CSV, FLIGHT_FIELDS)

def save_flights(flights):
    save_csv(FLIGHTS_CSV, FLIGHT_FIELDS, flights)

def get_seat_map(seats_string):
    return [list(row) for row in seats_string.split('_')]

def show_seat_map(seat_map):
    print("\nSEAT MAP (X=booked, letters=available):")
    header = "       " + " ".join(col for col in seat_map[0])
    print(header)
    print("     " + "-" * len(header))
    for i, row in enumerate(seat_map):
        rownum = f"Row {i+1:2}:"
        seats = " ".join(row)
        print(f"{rownum}  {seats}")
    print()

def choose_seat(seat_map):
    rows = len(seat_map)
    cols = len(seat_map[0])
    while True:
        try:
            row = int(input(f"Select row (1-{rows}): ")) - 1
            col_char = input(f"Select seat letter ({seat_map[0][0]}-{seat_map[0][-1]}): ").upper()
            col = ord(col_char) - ord(seat_map[0][0])
            if (0 <= row < rows) and (0 <= col < cols):
                if seat_map[row][col] != 'X':
                    return f"{row+1}{col_char}"
                else:
                    print("Seat already booked!")
            else:
                print("Invalid seat selection.")
        except:
            print("Invalid input.")

def search_flights():
    flights = load_flights()
    if not flights:
        print("No flights available!\n")
        return []
    print("\n-- FLIGHT SEARCH --")
    print("1. Search All Flights")
    print("2. Filter by Origin/Destination/Airline")
    print("3. Filter by Date")
    choice = input("Choose option (1-3): ")
    filtered = flights
    if choice == '2':
        orig = input("From (leave blank to skip): ").strip().lower()
        dest = input("To (leave blank to skip): ").strip().lower()
        airline = input("Airline (leave blank to skip): ").strip().lower()
        filtered = [
            f for f in flights
            if (not orig or f['origin'].strip().lower() == orig)
            and (not dest or f['destination'].strip().lower() == dest)
            and (not airline or f['airline'].strip().lower() == airline)
        ]
    elif choice == '3':
        dt = input("Date (YYYY-MM-DD or blank for all): ").strip()
        filtered = [f for f in flights if f['date'] == dt] if dt else flights
    if not filtered:
        print("\n  !! No flights found with your filter !!\n")
        return []
    print("\nAvailable Flights:")
    print("-" * 110)
    for f in filtered:
        print(f"ID: {f['flight_id']} | {f['airline']:9} | {f['origin']} → {f['destination']:9} | "
              f"{f['date']} {f['time']:>5} | ₹{f['price']}")
    print("-" * 110)
    return filtered

def payment_process(amount):
    print("\n=== PAYMENT GATEWAY ===")
    print(f"Amount To Pay: ₹{amount}")
    print("1. Card | 2. UPI | 3. Net Banking | 4. Wallet")
    method = input("Select payment method (1-4): ")
    _ = input("Enter payment details: ")
    print("Processing payment", end="", flush=True)
    for _ in range(3):
        print(".", end="", flush=True)
        time.sleep(0.45)
    print()
    if random.randint(1, 10) <= 9:
        print("Payment Successful! 🎉")
        payref = f"PY{random.randint(100000,999999)}"
        print(f"Payment Ref: {payref}\n")
        print("---------- RECEIPT ----------")
        print(f"Amount Paid: ₹{amount}")
        print(f"Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        print(f"Payment Reference: {payref}")
        print("-----------------------------\n")
        return True, payref
    else:
        print("Payment FAILED! Try again.\n")
        return False, None

def mark_seat_as_booked(seats_string, selected_seat):
    seat_map = get_seat_map(seats_string)
    row = int(selected_seat[:-1])-1
    col = ord(selected_seat[-1]) - ord(seat_map[0][0])
    seat_map[row][col] = 'X'
    return '_'.join([''.join(row) for row in seat_map])

def already_booked(flight_id, seat):
    bookings = load_csv(BOOKINGS_CSV, BOOKING_FIELDS)
    for b in bookings:
        if b['flight_id'] == flight_id and b['seat'] == seat:
            return True
    return False

def book_flight(username):
    flights = search_flights()
    if not flights:
        return
    flight_id = input("Enter Flight ID to book: ")
    flight = next((f for f in flights if f['flight_id'] == flight_id), None)
    if not flight:
        print("Flight not found.")
        return
    seat_map = get_seat_map(flight['seats'])
    show_seat_map(seat_map)
    while True:
        seat = choose_seat(seat_map)
        if already_booked(flight_id, seat):
            print("Sorry, this seat was just booked! Try another.")
        else:
            break
    paid, payref = payment_process(flight['price'])
    if not paid:
        print("Booking NOT completed.")
        return
    flight['seats'] = mark_seat_as_booked(flight['seats'], seat)
    all_flights = load_flights()
    for f in all_flights:
        if f['flight_id'] == flight['flight_id']:
            f['seats'] = flight['seats']
    save_flights(all_flights)
    bookings = load_csv(BOOKINGS_CSV, BOOKING_FIELDS)
    booking_id = f"BKG{len(bookings)+1:04}"
    bookings.append({
        'booking_id': booking_id,
        'username': username,
        'flight_id': flight['flight_id'],
        'seat': seat,
        'timestamp': datetime.now().isoformat(),
        'amount': flight['price'],
        'payment_ref': payref
    })
    save_csv(BOOKINGS_CSV, BOOKING_FIELDS, bookings)
    print(f"Booking successful! Your Booking ID: {booking_id}, Seat: {seat}, Airline: {flight['airline']}")

def view_my_bookings(username):
    bookings = load_csv(BOOKINGS_CSV, BOOKING_FIELDS)
    my_bookings = [b for b in bookings if b['username'] == username]
    if not my_bookings:
        print("No bookings found.\n")
        return
    print("\nYOUR BOOKINGS & HISTORY")
    print("-" * 110)
    total = 0
    for b in my_bookings:
        all_flights = load_flights()
        flight = next((f for f in all_flights if f['flight_id'] == b['flight_id']), None)
        airline = flight['airline'] if flight else "N/A"
        print(f"BookingID: {b['booking_id']} | Airline: {airline:9} | Flight: {b['flight_id']} | "
              f"Seat: {b['seat']} | ₹{b['amount']} | Paid Ref: {b['payment_ref']}")
        print(f"   On: {b['timestamp'][:19]}")
        total += int(b['amount'])
    print("-" * 110)
    print(f"TOTAL AMOUNT PAID: ₹{total}\n")

def cancel_booking(username):
    bookings = load_csv(BOOKINGS_CSV, BOOKING_FIELDS)
    my_bookings = [b for b in bookings if b['username'] == username]
    if not my_bookings:
        print("No bookings to cancel.\n")
        return
    booking_id = input("Enter Booking ID to cancel: ")
    booking = next((b for b in bookings if b['booking_id'] == booking_id and b['username'] == username), None)
    if not booking:
        print("Booking not found!")
        return
    flights = load_flights()
    flight = next((f for f in flights if f['flight_id'] == booking['flight_id']), None)
    if flight:
        seat_map = get_seat_map(flight['seats'])
        row = int(booking['seat'][:-1])-1
        col = ord(booking['seat'][-1]) - ord(seat_map[0][0])
        seat_map[row][col] = chr(ord(seat_map[0][0])+col)
        flight['seats'] = '_'.join([''.join(r) for r in seat_map])
        for f in flights:
            if f['flight_id'] == flight['flight_id']:
                f['seats'] = flight['seats']
        save_flights(flights)
    bookings.remove(booking)
    save_csv(BOOKINGS_CSV, BOOKING_FIELDS, bookings)
    print("Booking cancelled.\n")

def passenger_menu(username):
    while True:
        print("""
╔════════════════════════════════════════════════════╗
║             Flight Booking Main Menu              ║
╠════════════════════════════════════════════════════╣
║ 1. View/Search Flights                            ║
║ 2. Book a Flight Seat                             ║
║ 3. View My Bookings/History                       ║
║ 4. Cancel a Booking                               ║
║ 5. Logout                                         ║
╚════════════════════════════════════════════════════╝
        """)
        cmd = input("Choose option: ")
        if cmd == '1':
            search_flights()
        elif cmd == '2':
            book_flight(username)
        elif cmd == '3':
            view_my_bookings(username)
        elif cmd == '4':
            cancel_booking(username)
        elif cmd == '5':
            print("Logged out.\n")
            break
        else:
            print("Invalid choice.")

def main_menu():
    print("\n🔷 Welcome to Flight Booking System (Console Edition) 🔷\n")
    while True:
        print("""
============ MENU ============
1. Login
2. Register
3. Exit
==============================
""")
        choice = input("Choose option: ")
        if choice == '1':
            username = login_user()
            if username:
                passenger_menu(username)
        elif choice == '2':
            register_user()
        elif choice == '3':
            print("Goodbye!")
            break
        else:
            print("Invalid choice.")

if __name__ == "__main__":
    main_menu()
```
